```python
# Data Vis

import matplotlib.pyplot as graph

y = [1,3,8,2]
x = [1,2,3,4]

graph.plot(y,x)
graph.title("A Sample line graph")
graph.xlabel("X axis")
graph.ylabel("Y axis")
graph.show()
```

```python
import matplotlib.pyplot as graph
import matplotlib.style as style

style.use("ggplot")

y1 = [1,3,8,2]
x1 = [1,2,3,4]

y2 = [2,6,7,2]
x2 = [1,2,2,4]

graph.plot(y1,x1)
graph.plot(y2,x2)

graph.title("A Sample line graph")
graph.xlabel("X axis")
graph.ylabel("Y axis")
graph.show()
```

```python
import matplotlib.pyplot as graph
import matplotlib.style as style

style.use("ggplot")

y1 = [1,3,8,2]
x1 = [1,2,3,4]

y2 = [2,6,7,2]
x2 = [1,2,2,4]

graph.plot(y1,x1, label = "First Line", linewidth=5, color='r')
graph.plot(y2,x2, label = "Second Line", linewidth=5, color='b')

graph.title("A Sample line graph")
graph.xlabel("X axis")
graph.ylabel("Y axis")
graph.legend()
graph.show()
```

```python
import matplotlib.pyplot as graph
import matplotlib.style as style

style.use("ggplot")

y1 = [1,3,8,2]
x1 = [1,2,3,4]

y2 = [2,6,7,2]
x2 = [1,2,2,4]

graph.bar(y1,x1, label = "First Line", linewidth=1, color='r')
graph.bar(y2,x2, label = "Second Line", linewidth=1, color='b')

graph.title("A Sample line graph")
graph.xlabel("X axis")
graph.ylabel("Y axis")
graph.legend()
graph.show()
```
