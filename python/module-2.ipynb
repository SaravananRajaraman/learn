```python
# Collections

# list
# Its a mutable
# Its rep by []
# Its is indexed
# Its allow to duplicates the values
# Its is ordered

name = ['saravana', 'ram', 'liam', 'Vhin'];
print(name)

for x in name :
    print('Welcome ...', x)
    
# Add to list
## .append(<value>) 
## .insert(<position>, <value>)

name.append('raja')
print(name)

name.insert(0, 'arun')
print(name)

# Remove 
## .pop() //Last ele will removed 
## .remove(<value>)
## del Remove ele with index

name.pop()
print(name)

name.remove('ram')
print(name)

del name[len(name) - 1]
print(name)

# Copy
## .copy() //not mutable, its clone
## assign =

name2 = name.copy()
print(name2)

name3 = name
name.pop()

print(name3)

# Some methods
## Get the count of values
## .reverse() to reverse order
## .sort()

name.append('arun')
print(name.count('arun'))

name.reverse()
print(name)

name.sort()
print(name)

name.clear()
print(name)
```

```python
# Tuple
## Its a mutable
## Its rep by ()
## Its is indexed
## Its allow to duplicates the values
## Its is ordered

# Some methods
## Get the count of values
## .count(<value>)
## delete entire list (del)

name = ('saravana', 'ram', 'liam', 'Vhin');
print(name)
print(name.count('ram'))

del name
```

```python
# Set

## Its a mutable
## Its is unindexed
## Its is unordered
## Its doesnt allow duplicates the values
## Its rep by {}

name = {'saravana', 'ram', 'liam', 'Vhin'};

### The order will gen as unordered 
print(name)

for x in name:
    print(x)

# Add

name.add('kumar')

print(name)

for x in name:
    print(x)
    
# Pop will remove the first ele

name.pop()
print(name)

# Remove the values in set

name.remove('ram')
print(name)

name.discard('liam')
print(name)

del name

# some math stuff

## .union(<set>)
## .intersection(<set>)

a = {1,3,5,7}
b = {2,4,6,8,9,1}

print(a.union(b))
print(a.intersection(b))
print(a.issubset(b))
print(a.issuperset(b))
print(a.update(b))
print(a)
```

```python
# Dictinoary 

## It's a mutable
## It's a indexed
## It's dose not allow duplicate
## It's rep by { KEY : VALUE }
## It's a ordered or unordered

d1 = {
    'name':'one',
    'age': 28,
    'sex' : 'male',
    'location' : {
        'tn' : ['chennai', 'trichy'],
        'kn' : ['mysore', 'benglore']
    }
}

print(d1)
print(d1.keys())
print(d1.values())

for x in d1 :
    print(x, d1[x])

for x,y in d1.items() :
    print(x,y)
    
# Add

d1['id'] = "123432"
print(d1)

# Remove

d1.pop('location')
print(d1)

d1.popitem()
print(d1)

del d1['name']
print(d1)


```
